# Step 1: Initialize Population

```python
[i for i in range(4)]
```

```python
import random
def initPopulation(n,b):
    p={}
    for i in range(n):
        s=[]
        for j in range(b):
            s.append(random.choice([0,1]))    
        p[i]=s
    return p      

```

```python
population=initPopulation(8,5)
population
```

# Step 2: Parent Selection

Fitness Calculation

```python
[i for i in range(4,-1,-1)]
```

```python
def getFitness(p):
    f=[]
    for k in p.keys():
        d = 0
        b = len(p[k])
        for i in range(b-1,-1,-1):
            d=d+p[k][i]*2**(4-i)
        f.append(d**2)
    return f
```

```python
fitness = getFitness(population)
fitness
```

```python
total_fitness = sum(fitness)
prob = [round(f/total_fitness,3) for f in fitness]
prob
```

```python
def select_parent(n,probabilities):
    parents = {}
    for i in range(n):
        r =random.choice(range(1,1001,1))/1000
        s = 0
        for j in range(len(probabilities)):
            s = s + probabilities[j]
            if(s>r):
                parents[i]=population[j]
                #print(population[j])
                break
    return parents
```

```python
parents = select_parent(6,prob)
parents
```

# Step 3: Crossover

```python
def crossover(cp1,cp2,parents):
    for i in range(5):
        a=random.choice([0,1])
        if a==1:
            temp=parents[1][i]
            parents[1][i]=parents[5][i]
            parents[5][i]=temp
        return parents
```

```python
cp1=random.choice(range(0,6))
cp2=random.choice(range(0,6))

crossover(cp1,cp2,parents)
fitness=getFitness(parents)
total_fitness = sum(fitness)
prob = [round(f/total_fitness,3) for f in fitness]
parents = select_parent(6,prob)

parents
```
